```python
import numpy as np
months=np.array(["jan","feb","march","april","may","june","july","august","sep","october","november","dec"])
sales=[]
print("enter the sales (in $1000) for each month")
for month in months:
  value=float(input(f"{month}:"))
  sales.append(value)
sales=np.array(sales)

print("\n-------Company Sales anyalisis-----")
print("\n Total sales of the year: ",np.sum(sales),"$")
print("\n Average monthly sales : ",np.mean(sales))
print("\n Highest sales:",np.max(sales))
print("\n Lowest sales: ",np.min(sales))
best_month=months[np.argmax(sales)]
worst_month=months[np.argmin(sales)]

print("best Month: ",best_month,months[np.argmax(sales)])
print("worst Month: ",worst_month)

above_avg=months[sales>np.mean(sales)]
below_avg=months[sales<np.mean(sales)]

print("Above average months: ",above_avg)
print("below average months: ",below_avg)

print("\n how much each month differs from the average: ")
differnce=sales-np.mean(sales)
print(differnce)
```
